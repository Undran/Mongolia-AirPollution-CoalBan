```python
# eda_pm25.ipynb
# Step 0: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/Undran/Mongolia-AirPollution-CoalBan/refs/heads/main/data/ulaanbaatar_pm25_subset.csv")

# Step 2: Inspect the dataset
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
display(df.head())

# Step 3: Summary statistics
print("Summary statistics for numeric columns:")
display(df.describe())

# Step 4: Check for missing values
print("Missing values per column:")
display(df.isna().sum())

# Step 5: Categorical exploration
for col in ["parameter", "location", "city", "sourcename"]:
    if col in df.columns:
        print(f"Value counts for {col}:")
        display(df[col].value_counts().head(10))

# Step 6: Visualization
# Histogram of PM2.5 values
plt.figure(figsize=(10,5))
sns.histplot(df["value"], bins=30, color="red", kde=True)
plt.title("Distribution of PM2.5 values (top 5000 highest)")
plt.xlabel("µg/m³")
plt.show()

# Optional: Boxplot by location
plt.figure(figsize=(12,5))
sns.boxplot(data=df, x="location", y="value")
plt.xticks(rotation=90)
plt.title("PM2.5 values by location")
plt.show()
```
