```python
import pandas as pd
import os

# Ensure data/ folder exists
os.makedirs("data", exist_ok=True)

# Load dataset
df = pd.read_csv("/kaggle/input/ulaanbaatar-particulate-matter/Ulaanbaatar Particulate Matter Sensor Data.csv")


# List of Ger district stations
ger_stations = ['Bukhiin urguu', 'Nisekh', 'Amgalan', 'Baruun 4 zam', 
                'Mongol gazar', 'Tolgoit', 'Misheel expo', '100 ail', 
                'Urgakh naran', 'Bayankhoshuu']

# Filter for Ger district stations and keep first 2000 rows (optional)
df_subset = df[df['location'].isin(ger_stations)].head(2000)

# Check first 5 rows of first 5 columns
print(df_subset.iloc[:5, :5])

# Save subset
df_subset.to_csv("data/ulb_pm_subset.csv", index=False)
print("Subset saved as 'data/ulb_pm_subset.csv'")
```
